# Independent variable 

### Spending in 1000 dollar unit   (Continuous data)
### Card holds value 1 for having credit card and 0 for not having(Categorical data)

# Dependent variable 

### coupon used is 1 ; not used is zero (categorical data)


```python
ind_vars= df[['Spending','Card']]
ind_vars_f = sm.add_constant(ind_vars)
model = sm.Logit(df['Coupon'],ind_vars_f)
res = model.fit()
print(res.summary2())
```
