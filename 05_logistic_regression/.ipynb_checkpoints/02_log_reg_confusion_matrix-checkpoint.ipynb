```python
import pandas as pd
from sklearn.linear_model import LogisticRegression as LogReg
from sklearn.model_selection import train_test_split as tts
```

```python
df = pd.read_excel(r'D:\ERA_ML\IIT Roorkee\Dataset\week_8\Week 8 (Important Data)-20220425T104320Z-001\Week 8 (Important Data)\Simmons.xls')
df.head()
```

```python
X = df[['Spending','Card']]
Y = df['Coupon']
X_train,X_test, Y_train, Y_test  =  tts(X,Y,test_size=0.25,random_state=100)
print(X_train.shape)
```

```python
# To get help for the ML Algorithm 
LogReg?
```

```python
model = LogReg(solver='newton-cg')
fit_mod_1 = model.fit(X_train,Y_train)
fit_mod_1.coef_
```

```python
print(fit_mod_1.intercept_)
print(fit_mod_1.coef_)
```

# Cross verify the statsmodels result 


<h2 style ='color:red'> We see a difference between statsmodels to sklearn LogisticRegression Algorithm</h2>


```python
y_pred = fit_mod_1.predict(X_test)
y_pred
#y_pred.shape  : (25,)
```

# Validate how well the LogReg worked for us 

```python
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, roc_auc_score, precision_score
#confusion_matrix?
print(confusion_matrix(Y_test,y_pred))

# unpack each individual in confusion matrix 
# true_negative, false_positive, false_neg, true_pos 
tn, fp, fn, tp = confusion_matrix(Y_test,y_pred).ravel()
print(tn, fp, fn, tp)

# Sensitivity 
sensitivity = tp/(tp+fn)
print(f'The model we have trained is capable of predicting {sensitivity} coupon used when actually it is used')

#Specificity
specificity = tn /(tn+fp)
print(f'The model we have trained is capable of predicting {round(specificity,2)} coupon not used when actually it is not used')
```

```python
# Some thoughts about predict method and predict_proba method
# Predict method actually classifies either 0 or 1 based on the default threshold 0.5 
# predict_proba just returns as the probablity value for each entry 
# [0.27892448, 0.72107552]  --> 0.27 is the probablity for classifying it as x 0.72 for classifying it as y
# You can customise the threshold as you wish 


y_pred_proba = fit_mod_1.predict_proba(X_test)
print(type(y_pred_proba))
print(y_pred_proba.shape)
y_pred_proba
```

```python
# True positive rate  -- Sensitivity
# False Positive rate  -- (1-specificity)
from sklearn import metrics
import matplotlib.pyplot as plt
fpr = fp/(fp+tn)
print('fpr',fpr)

#define metrics
y_pred_proba = fit_mod_1.predict_proba(X_test)[::,1]  # passing only the prob of being 1 
print(y_pred_proba)
fpr, tpr, _ = metrics.roc_curve(Y_test,  y_pred_proba)

#create ROC curve
plt.plot(fpr,tpr)
plt.scatter(0.06,0.6,marker='*')  # This value is for the dafault threshold value 0.5 , It should lie on the ROC curve
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
