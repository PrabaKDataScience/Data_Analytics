# Simple Linear Regression

```python
x =np.column_stack((tbl1['AdvertisingExpenditure($1000s)'],tbl1['Price']))
y = tbl1['Sales(1000s)']
x2 = sm.add_constant(x)
model = sm.OLS(y,x2)
Model = model.fit()
print(Model.summary())
```

# Adding Interaction to the model

```python
z1 =tbl1['AdvertisingExpenditure($1000s)']
z2 = tbl1['Price']
z3 = z1*z2  # Interaction between predictors z1, z2
```

```python
x_new =np.column_stack((z1,z2,z3))
y = tbl1['Sales(1000s)']
xnew2 = sm.add_constant(x_new)
model2 = sm.OLS(y,xnew2)
Model2 = model2.fit()
print(Model2.summary())
```
