```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
tbl1 = pd.read_excel(r'D:\ERA_ML\IIT Roorkee\Dataset\week_9\Week 9 Important Data-20220425T104519Z-001\Week 9 Important Data\MPG.xlsx')
tbl1
```

```python
plt.scatter(tbl1['Weight'],tbl1['MilesperGallon'])
plt.ylabel('MilesperGallon')
plt.xlabel('Weight')
```

```python
x =tbl1['Weight']
y = tbl1['MilesperGallon']
x2 = sm.add_constant(x)
model = sm.OLS(y,x2)
Model = model.fit()
print(Model.summary())
```

```python
E=Model.resid_pearson
E
```

```python
yhat = Model.predict(x2)
yhat
```

```python
plt.scatter(yhat,E)
```

#### The residual plot has heteroscedasticity (Conic shape) . This violates the assumption of linear regression !!!

Image is just to understand the pattern . Nothing to do with the data we are dealing 




```python
Y = np.log(y)
```

```python
model2 = sm.OLS(Y,x2)
Model2 = model2.fit()
print(Model2.summary())
```

```python
E2=Model2.resid_pearson
E2
```

```python
yhat = Model2.predict(x2)
yhat
```

```python
plt.scatter(yhat,E2)
```
