```python
import pandas as pd

data = {'bill':[34,108,64,88,99,51],'tip':[5,17,11,8,14,5]}

df =pd.DataFrame(data)
df
```

```python
# Data visualisation
import seaborn as sns
my_plot = sns.scatterplot(data=df, x="bill", y="tip",)
my_plot.set_title('Bill Vs Tip',fontsize = 20)
my_plot.set_xlabel("Bill ($)", fontsize = 15)
my_plot.set_ylabel("Tip ($)", fontsize = 15)
# BEST PRACTICE
# Have the independent variable in the X -axis 
# Have the dependent variable in the y axis 
```

```python
# Do a Regplot to find the relationship
sns.regplot(data=df, x="bill", y="tip")
```

```python
df.corr()  # It actually produces a mirror matrix about the diagonal

# There is strong positive relationship between Bill and Tip    
# Bill is propotional to Tip 
```

```python
# Descrptive statistics summary
df.describe()
```

```python
# Calculating the slope
X=df['bill']
x_bar = X.mean()
y=df['tip']
y_bar = y.mean()
centroid_point = (x_bar,y_bar)
print('Centroid is :', centroid_point)

slope_num = 0
slope_den = 0

for i in range(len(df['bill'])):
    x_covar = df['bill'][i]-x_bar
    #print(x_covar)
    y_covar = df['tip'][i]-y_bar
    #print(y_covar)
    covar = x_covar*y_covar
    #print(covar)
    slope_num+=covar
    x_var = x_covar**2
    #print(slope_den)
    slope_den+=x_var
    #print('*****')


slope = slope_num/slope_den
print('The slope is ',slope)
    
    
```

```python
# Calculating the y intercept

intercept = y_bar-slope*x_bar
print('The intercept is ',intercept)
```

# Validating the slope and intercept using sklearn 

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Reshape the Independent variable like 6 rows and 1 column
arr = np.array(X,dtype='int8')
#print(arr)
X =arr.reshape(-1, 1)
#print(X)
print('The shape of the independent variable or Predictor is :', X.shape)

y= np.array(df['tip'])

# Fiting to the model
reg = LinearRegression().fit(X,y)



#Unpacking the result
print('The slope is :', reg.coef_)
print('The intercept is :',reg.intercept_)

#Getting the coefficient of determination 
print('The coefficient of determination :',reg.score(X, y))
```
