```python
import pandas as pd
import statsmodels.formula.api as sfa
from statsmodels.stats.anova import anova_lm
```

```python
df  = pd.read_excel(r'D:\ERA_ML\IIT Roorkee\Dataset\week_7\Week 7 (Important Data)-20220425T104054Z-001\Week 7 (Important Data)\icecream.xlsx')
df
```

```python
reg = sfa.ols('Sales ~Student_Population',data=df).fit()
reg.summary()
```

```python
anova_lm(reg)
```

```python
import seaborn as sns
x= df['Student_Population']
y= df['Sales']  # Though we pass Sales here, since we plot Residual plot , it calcualtes the residualof sales for the y value
plt = sns.residplot(x, y, color="g")
plt.set_ylabel('Residual of Sales')
```

```python
#Standarized residual plot 

# We can make use of the earlier regression model itself
influence = reg.get_influence() 
resid_student = influence.resid_studentized_external
resid_student
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(x,resid_student, color = "green")
```

# Interpretation

The values are between -2 to 2 , hence the assumption of normality of error term is normal distributed

```python
# Normal probablity Plot NPP
import statsmodels.api as s
from scipy import stats
res = reg.resid # residuals
probplot = s.ProbPlot(res,stats.norm, fit=True)
fig = probplot.qqplot(line='45')
h = plt.title('Ex. 1 - qqplot - residuals of OLS fit') 
plt.show()
```
