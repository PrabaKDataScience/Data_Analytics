```python
import pandas as pd
df = pd.read_excel("D:\ERA_ML\IIT Roorkee\Dataset\week_5\cotton weight.xlsx")
df
```

```python
new_df = pd.melt(df.reset_index(),id_vars=['index'])
new_df.rename(columns={'variable':'treatments'},inplace=True)  # Specific column renaming
new_df.head(10)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model_1 = ols('value ~C(treatments)',data=new_df).fit()
model_1.summary()
```

```python
anova_table = sm.stats.anova_lm(model_1,typ=1)
anova_table
```

```python
# Understanding:
# The more the F value --> lesser the P value --> High chances to reject null hypothesis
# We reject null hypothesis since P < aplha 
#dir(mod_1)
f_val_from_test = model_1.fvalue
p_val_from_test = model_1.f_pvalue
alpha = 0.05
if p_val_from_test<=alpha:
    print('Reject Null hypothesis')
else:
    print('Fail to reject null hypothesis')
```

```python
#POSTHOC ANALYSIS
# We Reject Null hypothesis 
# that doesn't mean all means are unequal
# Either any to pairs may be equal
# Lets test using Tukey cramer test 

from statsmodels.stats.multicomp import pairwise_tukeyhsd
# perform Tukey's test
tukey = pairwise_tukeyhsd(endog=new_df['value'],
                          groups=new_df['treatments'],
                          alpha=0.05)

#display results
print(tukey)
```
