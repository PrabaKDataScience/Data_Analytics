# RBD - Intro


When the dependent variable is depend only on the treatment(independent var)  We do Simple Randomised design or Completely randomised  Design CRD

<h3>Blocking Variable</h3>
<p>In a&nbsp;<a href="https://stattrek.com/statistics/dictionary.aspx?definition=randomized-block-experiment" rel="external">randomized block experiment</a>, a good blocking variable has four distinguishing characteristics:</p>
<ul>
    <li>It is included as a factor in the experiment.</li>
    <li>It is not of primary interest to the experimenter.</li>
    <li>It affects the dependent variable.</li>
    <li>It is unrelated to independent variables in the experiment.</li>
</ul>
<p>A blocking variable is a potential nuisance variable - a source of undesired variation in the dependent variable. By explicitly including a blocking variable in an experiment, the experimenter can tease out nuisance effects and more clearly test treatment effects of interest.</p>

When the dependent variable having impact of blocking variable , we need to create block and do Randomised Block Design RBD




Ref : https://discovery.cs.illinois.edu/learn/Basics-of-Data-Science-with-Python/Experimental-Design-and-Blocking/

```python
# The stress level(Dependent variable) is depend on System Type (Independent variable)
# Also the stress level(Dependent variable) is having impact because of blocking variable(Controller who works)
# We are not interested to know which system works well for each controller , 
# rather we are interested to know which system works for all 
# If I dont consider the controller ability i.e No block then it is CRD completely random design

import statsmodels.formula.api as sfa
import statsmodels.api as sm
import pandas as pd

df  = pd.read_excel(r'D:\ERA_ML\IIT Roorkee\Dataset\week_6\Important Data-20220425T103808Z-001\Important Data\RBD.xlsx')

alpha = 0.05

new_df = pd.melt(df.reset_index(),id_vars=['index'])
#print(new_df)
new_df.columns=['index','treatments','values']
#print(new_df)
model_1 = sfa.ols('values ~C(treatments)',new_df).fit()
model_1.summary()


    
```

```python
anova_table = sm.stats.anova_lm(model_1,typ=1)
anova_table
```

```python

f_val_from_test = model_1.fvalue
p_val_from_test = model_1.f_pvalue
alpha = 0.05
if p_val_from_test<=alpha:
    print('Reject Null hypothesis')
else:
    print('Fail to reject null hypothesis')
```

<b><i> CRD - This implies that three system does not have any significant differences in the mean at alpha 0.05 
    
 Note that SSE here is 49 </i></b>

```python
# RANDAMISED BLOCK DESIGN RBD 
# Including the controller ability  

import statsmodels.formula.api as sfa
import statsmodels.api as sm
import pandas as pd

df  = pd.read_excel(r'D:\ERA_ML\IIT Roorkee\Dataset\week_6\Important Data-20220425T103808Z-001\Important Data\RBD.xlsx')

alpha = 0.05

new_df = pd.melt(df.reset_index(),id_vars=['index'])
#print(new_df)
new_df.columns=['block','treatments','values']
#print(new_df)
model_2 = sfa.ols('values ~C(block)+C(treatments)',new_df).fit()
model_2.summary()
```

```python
anova_table = sm.stats.anova_lm(model_2,typ=1)
anova_table
```

```python

f_val_from_test = model_2.fvalue
p_val_from_test = model_2.f_pvalue
alpha = 0.05
if p_val_from_test<=alpha:
    print('Reject Null hypothesis')
else:
    print('Fail to reject null hypothesis')
```

<b><i> RBD - This implies that three system does have any significant differences in the mean at alpha 0.05 
    
 Note that SSE here is 19 reduced by 30(CRD) </i></b>
