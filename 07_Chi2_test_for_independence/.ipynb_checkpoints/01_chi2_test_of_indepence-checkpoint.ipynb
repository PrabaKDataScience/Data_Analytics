<p>The Chi-square test of independence checks whether two variables are likely to be related or not. We have counts for two categorical or nominal variables. We also have an idea that the two variables are not related. The test gives us a way to decide if our idea is plausible or not.</p>

```python
from scipy.stats import chi2_contingency
import pandas as pd 
```

```python
data = pd.read_excel(r'D:\ERA_ML\IIT Roorkee\Dataset\week_10\Week 10 (Important Data)-20220425T104557Z-001\Week 10 (Important Data)\acad.xlsx')
data.head()
```

```python
## Explanation of the dataset ##
# Rsp No -- Student roll no 
# aa - Academic ability - Test score out of 100 -- Interval data 
# pe - Parent Education - Its actually a categorical but we took survey to be as categorical 
# sm - Student motivation 
    # 0 - if motivated to spend extra time to get good marks 
    # 1 - Student not yet decide
    # 2 -  Decided to spend extra time
    
# ae - Academic Evaluation - Teachers intuition about the student performance 
# r - religion 
# g - Gender (0 for male , 1 for female)
# community 


# The dependent variable is aa and all the other columns/ selected columns can be independent value or predictors 

# We select two independent values or predictors from all the availabel and test whether they are 
#actually dependent or independent based on the data 

# ** Come up with a conclusion that there is a dependency between two categorical/nomial data or not **



```

```python
# Though we have many columns , we are interested with only examing 
# Whether the gender and sm are dependent or independent

### Hypothessis test ###

# H0: The gender and sm are independent 
# Ha : The gender and sm are not independent 
```

```python
con_table =  pd.pivot_table((data[['g','sm']]), index='g', columns='sm', aggfunc=len)
con_table
```

```python
chi2,p,dof,exp_vals = chi2_contingency(con_table)
```

```python
chi2
```

```python
p
```

```python
# Since P value is above Significance level(0.05), We accept null hypothesis

## The gender and sm are independent ##
```
