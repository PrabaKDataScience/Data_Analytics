# Sampling distribution of sampling mean

```python
# We will verify whether we get the mu as 275(Hypo testing) from a random sample of size 20
# We know that stock market data has high dependency in the latest price than it was in the past.
#However we are just doing it for exploration

# Find the sample mean --> Infer the population mean (Hypothesis test - 1 sample)
import numpy as np

sample_size =20 
number_of_samples =200
rs = []
np.random.seed(20)  # Seeding so that each run will not change the resulting random result
for i in range(number_of_samples):
    rs.append(np.random.choice(new_df['Close'], sample_size))
rs = np.array(rs)

# Samples_taken
print(rs[:5] ) #show only the first five sample  # Each sample contains five(sample_size) picks from the population 

# Sample_mean
sampling_mean_list = rs.mean(axis=1)
sample_stdev=sampling_mean_list.std()
sample_mean = sampling_mean_list.mean()
#print(sampling_mean_list[:2] )
```

# Hypothesis testing

```python
# Hypothesis testing
#H0: mu = 277
#Ha: mu !=277
# Its two tailed  - 
# Even though we know population standard deviation ,
# I am going to perform t test (Like I dont have any idea about population standard deviation)

alpha = 0.05 
hypo_mean =  275.4 # Assuming population mean may be hypo mean 

# Step 1 : Define Hypothesis
# H0: mu<=10
# Ha: mu >10

# Step 2 : Define significance level 
# alpha : 0.05 

# step 3 : Decision control 
# t_test_statistics < -t_critical   -- Reject Null  
# t_test_statistics > t_critical   -- Reject Null 


```

```python
# get t_critical 
from scipy.stats import t
dof = (number_of_samples-1)
t_critical  = t.ppf(alpha,dof)  
print('t critical: ',t_critical)


#step 4 : Calculate test t statistics 

#method1 : Own calculation using theoriticl formulas

import math 
def find_t_test_statistics(samp_mean, hypo_mean, sam_std_dev, sample_size):
    standard_error = sam_std_dev/math.sqrt(sample_size)  # Standard deviation of sampling distribution of mean 
    return (samp_mean-hypo_mean)/standard_error

t_test_static = find_t_test_statistics(samp_mean=sample_mean, hypo_mean=hypo_mean,
                                       sam_std_dev=sample_stdev, sample_size=number_of_samples) 
print('t_test_static manual : ', t_test_static)

# Find P value for the t_test_statistics
'''
Left-tailed test: p-value = cdf(x)
Right-tailed test: p-value = 1 - cdf(x)
Two-tailed test: p-value = 2 * min{cdf(x) , 1 - cdf(x)}'''

# Its two tailed test 
p_value_1  = t.cdf(t_test_static,199)  
p_value_2  = 1-t.cdf(t_test_static,199)
p_value = 2*min(p_value_1,p_value_2)
print('p_value for the test statistics is : ', p_value)


from scipy import stats
t_test_stat_scipy = stats.ttest_1samp(sampling_mean_list,popmean=hypo_mean)

print('t_test_static scipy: ', t_test_stat_scipy.statistic)
print('t_test_static P value: ', t_test_stat_scipy.pvalue)
```

```python

# Step 5 : Statistical condition 
### 
if t_test_stat_scipy.pvalue<=alpha/2:
    print('Reject Null hypothesis')
    print(f'The population mean is not equal to {hypo_mean}')
else:
    print('Fail to reject Null')
    print(f'The population mean is equal to {hypo_mean}')
    
```
