Unlike the samplinng distribution of sample mean and amplinng distribution of sample proportion , the sampling distribution of sample variances doesnt follow the Normal distribution, Its F distribution.

```python
#Given 
n1=11;sample_std_dev_1=5.1
n2=16;sample_std_dev_2=4.7
significance=0.1


#H0: pop_var_1 = pop_var_2
#Ha: pop_var_1 != pop_var_2

from scipy.stats import f
import numpy as np

F_score = (5.1**2)/(4.7**2)
print(F_score)
right_critical_f_value=f.ppf((significance/2),n2-1,n1-1)
right_test_std_dev = np.sqrt(F_score*right_critical_f_value)
left_critical_f_value=f.ppf(1-(significance/2),n2-1,n1-1)
left_test_std_dev = np.sqrt(F_score*left_critical_f_value)

print(right_test_std_dev,left_test_std_dev)

#Since variance_interval has 1 , there is possibility that pop_var_1/pop_var_2=1 i.e pop_var_1 = pop_var_2
# We are fail to reject NULL Hypothesis
```
