# Hypothesis testing - Two tailed Z-test



<h1>Method 1 : Critical z value approach </h1>




```python
# Given 

n = 30 # Sample_size
x_bar = 505 # sample_mean
sigma = 10 # pop_std_dev
mu_0 = 500 # hypothesised mean 
alpha =0.03 # level of significance
```

<h3> Method 1 : Critical z value approach </h3>


<h3> Step 1 - Develop hypothesis </h3>
    

H0 : mu = 500 achieved 

Ha : mu !=500 Stop the process of packing (If less customer will be disappointed - More manufacturer cost of making increases)


<h3> Step 2 - Specify the level of significance</h3>

alpha = 0.03 or 3%

<b> Since population standard deviation is given we can go for Z-test 
    Based on the hypothesis we need to two tail test </b>
    



    



```python
# Step 3 - Decision statement
# Step 3 --> Get critical Z value 
# Z ctitical value from Z table for alpha/2 =  --> 2.17
from scipy.stats import norm
z_critical = round(norm.ppf(0.015),3)
z_critical
```

<h3>Step 3: Decision statement </h3>

test_statitics should be within -2.17 to 2.17 --> Fail to reject null hypothesis



```python
# Step 4 : Calculate test statistics
import math 

def find_z_test_statistics(samp_mean, hypo_mean, pop_std_dev, sample_size):
    standard_error = pop_std_dev/math.sqrt(sample_size)  # Standard deviation of sampling distribution of mean 
    return (samp_mean-hypo_mean)/standard_error

z_test_static = find_z_test_statistics(samp_mean=x_bar, hypo_mean=mu_0, pop_std_dev=sigma, sample_size=n) 
z_test_static  
```

<h3> Step 5 : State Statistical condition: </h3>

Since the minus_z_alpha_by_2< z_test_static< plus_z_alpha_by_2--> <b>FAIL TO REJECT NULL<b>
    z_test_static = 2.73 which is not inside the acceptance region of NULL HYPOTHESIS 
    
    REJECT NULL 

```python
if -(abs(z_critical))<z_test_static<abs(z_critical): # Two  tail z test rule for fail to reject of null hypothesis
    print('Fail to reject Null Hypothesis')
else:
    print('Reject Null Hypothesis')
```

<h1> 
Method 2 : P value approach </h1>

<h3> Step 1 - Develop hypothesis </h3>
    

H0 : mu = 500 achieved 

Ha : mu !=500 Stop the process of packing (If less customer will be disappointed - More manufacturer cost of making increases)


<h3> Step 2 - Specify the level of significance</h3>

alpha = 0.03 or 3%

<b> Since population standard deviation is given we can go for Z-test 
    Based on the hypothesis we need to two tail test </b>
    

<h3>Step 3: Decision statement </h3>

test_statitics should be within -2.17 to 2.17 --> Fail to reject null hypothesis






<h3> Step 4 : Test statistics Calculation </h3>

```python
import math 

def find_z_test_statistics(samp_mean, hypo_mean, pop_std_dev, sample_size):
    standard_error = pop_std_dev/math.sqrt(sample_size)  # Standard deviation of sampling distribution of mean 
    return (samp_mean-hypo_mean)/standard_error

z_test_static = find_z_test_statistics(samp_mean=x_bar, hypo_mean=mu_0, pop_std_dev=sigma, sample_size=n) 
z_test_static  
```

```python
# Find the p value 
from scipy.stats import norm

p_value_one_side = 1-norm.cdf(z_test_static)

p_value_one_side

# Since it is a two tail test

p_value = p_value_one_side *2 

p_value
```

<h3> Step 5: Statistical cndition </h3>

p<=alpha -->  Reject null hypothesis


```python
if p_value<=alpha:
    print('Reject Null hypothesis')
    
else:
    print("Fail to reject null hypothesis")
```

<b style='color:green'>The packaging process should be stopped since the mean filling quantity is not 500ml . <b>

```python
# CONFIDENCE INTERVAL APPROACH

# confidence_interval for population mean = sample_mean +- (Reliability factor @ Zalpha_by_2)(Standard error)
upper_limit_confidence_interval = x_bar+(norm.ppf(alpha/2))*(sigma/sqrt(n))
lower_limit_confidence_interval = x_bar-(norm.ppf(alpha/2))*(sigma/sqrt(n))
print(lower_limit_confidence_interval, upper_limit_confidence_interval )  # 508.9620248387599 501.0379751612401


# The milk cartons are filling between 501 to 508 --> Excess of 1 ml to 8 ml 
# It is additional cost for company 
# Hence the filling should be stopped and corrected 

# The hypothesised mean should lie in the range --> 500 is not in the range 501 to 508 . Hence REJECT NULL HYPOTHESIS
```
