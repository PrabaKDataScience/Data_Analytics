```python
# Given 
# Formulation 1 : Standard chemistry 
# Formulation two : New drying ingredient

### Expected that Formulation two will have less drying time 
sigma_1 = sigma_2 = 8 
n_1 = n_2 =10
x1_bar = 121 
x2_bar = 112 
alpha = 0.05 

# H0 :  mu_1=mu_2 Drying time is same for both formulations
# Ha : mu_1 != mu_2 

from math import sqrt
from scipy.stats import norm

standard_error = sqrt((sigma_1**2/n_1)+(sigma_2**2/n_1))
print(standard_error)

Z_test_statistic = (x1_bar-x2_bar) / standard_error
print('Z_test_statistic', Z_test_statistic)

Z_critical = norm.ppf(alpha/2)
print(Z_critical)  # this value is -Z_alpha_by_2 , the +Z_alpha_by_2 also there

#Since Z_test 2.51 is above the Z_critical 1.96 , Reject null hypothesis
```
