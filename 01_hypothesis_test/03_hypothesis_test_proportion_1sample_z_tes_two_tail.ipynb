```python
# Given 
n= 120 
p_bar = 67/120 # Accident caused by drinking (SUCCESS factor in our problem)
q = 1-(67/120) # Accident caused not by drinking (FAILURE factor in our prob)
p_0 = 0.5 # Police claim that 50% of accident is due to drink and drive
alpha = 0.05 # Level of significance
```

<h3>Step 1 : Define hypothesis</h3>

H0: p = 0.5  
Ha: p != 0.5

<h3>Step 2 : Level of significance</h3>
alpha = 0.05

<h3>Step 3 : Decision statements</h3>






<h3> Step 4 : test statistics </h3>



```python
# Step 4 : Calculate test statistics
from math import sqrt
def get_z_proportion_score(sample_proportion,hypothesised_proportion,sample_size):
    standard_error = sqrt((hypothesised_proportion*(1-hypothesised_proportion))/sample_size)
    return (sample_proportion-hypothesised_proportion)/standard_error

z_test_static = get_z_proportion_score(sample_proportion=p_bar,hypothesised_proportion=p_0,sample_size=n)
print('z_test_static',z_test_static)
```

```python
# Find the p value
from scipy.stats import norm
p_value = (1-norm.cdf(z_test_static ))*2
print('p_value :  ',p_value)
```

```python
# Step 5 : Statistical condition

if p_value<=alpha:
    print('Reject Null hypothesis')
    
else:
    print("Fail to reject null hypothesis")
```

```python
# Using STATSMODEL library 

# Given 
# n= 120 
# p_bar = 67/120 # Accident caused by drinking (SUCCESS factor in our problem)
# q = 1-(67/120) # Accident caused not by drinking (FAILURE factor in our prob)
# p_0 = 0.5 # Police claim that 50% of accident is due to drink and drive
# alpha = 0.05 # Level of significance

from statsmodels.stats.proportion import proportions_ztest
# can we assume anything from our sample
significance = 0.05
sample_success = 67
sample_size = 120
# our Ho is  50%
null_hypothesis = 0.50
# check our sample against Ho for Ha > Ho
# for Ha < Ho use alternative='smaller'
# for Ha != Ho use alternative='two-sided'
stat, p_value = proportions_ztest(count=sample_success, nobs=sample_size, value=null_hypothesis, alternative='two-sided')
# report
print('z_stat: %0.3f, p_value: %0.3f' % (stat, p_value))
if p_value > significance:
   print ("Fail to reject the null hypothesis - we have nothing else to say")
else:
   print ("Reject the null hypothesis - suggest the alternative hypothesis is true")
```
