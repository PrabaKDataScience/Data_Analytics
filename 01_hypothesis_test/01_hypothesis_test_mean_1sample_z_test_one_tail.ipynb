# Hypothesis testing - One tailed Z-test

```python
n = 30 # Sample size
x_bar= 32 # smaple mean
sigma = 10 # population std dev
alpha =0.05 # level of significance
mu_0=30 # Hypothesied mean 

```

<h2 style="background-color:DodgerBlue;"> Method 1 : Critical value <h2>

<h3> Step 1 - Develop hypothesis </h3>
    

H0 : mu <= 30 achieved 

Ha : mu >30 Action required by delivery services director

<h3> Step 2 - Specify the level of significance</h3>

alpha = 0.05 or 5%

<b> Since population standard deviation is given we can go for Z-test 
    Based on the hypothesis we need to one tail test </b>
    
<h3> Step 3 - Decision statement</h3>
 Z ctitical value from Z table for alpha =  --> 1.64 
 
 Hence the test static should be below 1.64 so that we fail to reject Null hypothesis
 Else we reject Null and choose alternative hypothesis since data supports that 


    
    
   



```python
# OTHER WAY TO GET CRITICAL SCORE 
from scipy.stats import norm
z_critical = round(norm.ppf(0.025),3)
z_critical
```

<h3> Step 4 - Calculate test statistic </h3>





```python
import math 

def find_z_test_statistics(samp_mean, hypo_mean, pop_std_dev, sample_size):
    standard_error = pop_std_dev/math.sqrt(sample_size)  # Standard deviation of sampling distribution of mean 
    return (samp_mean-hypo_mean)/standard_error

z_test_static = find_z_test_statistics(samp_mean=x_bar, hypo_mean=mu_0, pop_std_dev=sigma, sample_size=n) 
z_test_static           

# n = 30 # Sample size
# x_bar= 32 # smaple mean
# sigma = 10 # population std dev
# alpha =0.05 # level of significance
# mu_0=30 # Hypothesied mean     
```

<h3> Step 5 : State Statistical condition: </h3>

Since the z_test_static<1.64 --> <b>FAIL TO REJECT NULL<b>
   

```python
if z_test_static>z_critical: # Right  tail z test rule for rejection of null hypothesis
    print('Fail to reject Null Hypothesis')
else:
    print('Reject Null Hypothesis')
```

<h2 style="background-color:DodgerBlue;"> Method 2 : P value method <h2>

<h3> Step 1 - Develop hypothesis </h3>
    

H0 : mu <= 30 achieved 

Ha : mu >30 Action required by delivery services director

<h3> Step 2 - Specify the level of significance</h3>

alpha = 0.05 or 5%

<b> Since population standard deviation is given we can go for Z-test 
    Based on the hypothesis we need to to tail test </b>
    
<h3> Step 3 - Decision statement</h3>
 
 Critical P vlaue is alpha 0.05 
 
if P_value of z_test_static is less than or eaual to Critical P value --> Reject Null hypothesis
 
We can reject null hypothesis if P_value<= alpha (level of significance)





    

<h3> Step 4 - Calculate test statistic </h3>





```python
import math 
from scipy.stats import norm

# n = 30 # Sample size
# x_bar= 32 # smaple mean
# sigma = 10 # population std dev
# alpha =0.05 # level of significance
# mu_0=30 # Hypothesied mean  

def find_z_test_statistics(samp_mean, hypo_mean, pop_std_dev, sample_size):
    standard_error = pop_std_dev/math.sqrt(sample_size)  # Standard deviation of sampling distribution of mean 
    return (samp_mean-hypo_mean)/standard_error

z_test_static = find_z_test_statistics(samp_mean=x_bar, hypo_mean=mu_0, pop_std_dev=sigma, sample_size=n) 
z_test_static           


# Find P value for the z_test_static 
# norm.cdf(z_test_static)  # Cumulative distribution function 
p_value_test_statistics = 1- norm.cdf(z_test_static)
p_value_test_statistics

print('Z value is {}. P value is {} '.format(z_test_static, p_value_test_statistics))


   
```

<h3> Step 5 : State Statistical condition: </h3>

Since the p_value_test_statistics>0.05 --> <b>FAIL TO REJECT NULL<b>
   

```python
if p_value_test_statistics>alpha: # Right  tail z test rule for rejection of null hypothesis
    print('Fail to reject Null Hypothesis')
else:
    print('Reject Null Hypothesis')
```
