```python
from sklearn.linear_model import LinearRegression as LR 
import pandas as pd
# import statsmodels.formula.api as sfa
# import statsmodels.api as sm
# import statsmodels.regression.linear_model as smlm
# import seaborn as sns
```

```python
df = pd.read_excel(r'D:\ERA_ML\IIT Roorkee\Dataset\week_7\Week 7 (Important Data)-20220425T104054Z-001\Week 7 (Important Data)\dummy.xlsx')
df
```

```python
new_df = pd.get_dummies(data=df,drop_first=False)
new_df
```

```python
X=new_df[['months_since_last_service','type_of_repair_electrical']]
Y=new_df['repair_time_in_hours']
```

```python
model=LR()
model.fit(X,Y)
print('Co ef are ', model.coef_)
print('Intercept', model.intercept_)
```

```python
coeff_parameter = pd.DataFrame(model.coef_,X.columns,columns=['Coefficient'])
coeff_parameter
```
